```python
# combine all the tables together, given they have same column name

majortableDFlist = []
checkList = ['Section', 'Date & Time', 'Room', 'Instructor', 'Quota', 'Enrol', 'Avail', 'Wait', 'Remarks']
for majorindex in range(len(tableMajorList)):
    for tableindex in range(len(tableMajorList[majorindex])):
        dftest = tableMajorList[majorindex][tableindex]
        dftest.columns = dftest.iloc[0]
        dftest = dftest[1:]
#         print(dftest.columns)
#         if len(dftest.columns) == len(checkList):
        if list(dftest.columns) == checkList:
            majortableDFlist.append(dftest)
```

```python
# majortableDFlist[1]
concat_df = pd.concat(majortableDFlist)
```
